```python
#健康データと睡眠データの統合　20251125

import pandas as pd
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter

# ============================================
# 1) 不要な日付列（J列など）を削除
# ============================================
# すべての列名から「日付」を検索し、A列以外の「日付」を削除
date_columns = [col for col in merged.columns if "日付" in col]

# A列の「日付」は必ず先頭なので残す
main_date_col = "日付"

for col in date_columns:
    if col != main_date_col:
        merged.drop(columns=[col], inplace=True)

# ============================================
# 2) 睡眠時間の小数1桁化
# ============================================
if "睡眠時間（時間）" in merged.columns:
    merged["睡眠時間（時間）"] = (
        pd.to_numeric(merged["睡眠時間（時間）"], errors="coerce").round(1)
    )

# ============================================
# 3) Excel出力
# ============================================
out_path = r"C:\Users\spax2\Documents\PythonWork\ExcelDATA\健康_睡眠_統合データ.xlsx"
merged.to_excel(out_path, index=False)

# ============================================
# 4) Excelの列幅・表示形式を整える
# ============================================
wb = load_workbook(out_path)
ws = wb.active

# A列（日付）の列幅
ws.column_dimensions["A"].width = 12

# A列の表示形式を日付(yyyy/mm/dd)に
for cell in ws["A"][1:]:
    cell.number_format = "yyyy/mm/dd"

# 睡眠時間（時間）の列を小数1桁に統一（Excel上でも）
header = [c.value for c in ws[1]]
if "睡眠時間（時間）" in header:
    idx = header.index("睡眠時間（時間）") + 1
    col_letter = get_column_letter(idx)
    for cell in ws[col_letter][1:]:
        cell.number_format = "0.0"

# ============================================
# 列タイトル（ヘッダー）を小さくして左寄せにする
# ============================================
from openpyxl.styles import Font, Alignment

for cell in ws[1]:  # 1行目がヘッダー
    cell.font = Font(size=10)                # 小さめのフォントに変更
    cell.alignment = Alignment(horizontal="left")  # 左寄せ

wb.save(out_path)
print("✨ 整形済みExcelを保存しました：", out_path)
```

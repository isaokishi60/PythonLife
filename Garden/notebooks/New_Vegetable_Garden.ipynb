```python
import pandas as pd
import openpyxl
import datetime
from openpyxl import utils

from openpyxl import load_workbook
filepath= r"C:\Users\spax2\Documents\PythonWork\農作業\vegetable_garden_photo_ex2.xlsx"
wb=load_workbook(filename=filepath)
ws1=wb['vegetable_garden_photo']
ws2=wb['Selected_Data']

# データ表のリセット
wb.remove(wb['Selected_Data'])
s2=wb.create_sheet('Selected_Data')

df=pd.read_excel(filepath, sheet_name = "vegetable_garden_photo", index_col="Date")
df.reset_index(inplace=True)

df
```

```python
import date_input_module
import pandas as pd

# 日付入力
print("開始日を入力してください")
start_date = date_input_module.validate_date(date_input_module.input_date())
print(start_date)
print("終了日を入力してください")
end_date = date_input_module.validate_date(date_input_module.input_date())
print(end_date)

# 品名入力
item_name = input("品名（Name or Item）を入力してください(カタカナ)：").strip()

# --- ここがポイント：DataFrame側を datetime64[ns] に統一 ---
df['Date'] = pd.to_datetime(df['Date'])        # .dt.date は付けない

# 入力側も Timestamp に
start_ts = pd.to_datetime(start_date)
end_ts   = pd.to_datetime(end_date)

# 抽出
if pd.notna(start_ts) and pd.notna(end_ts) and item_name:
    df2 = df[
        (df['Date'] >= start_ts) &
        (df['Date'] <= end_ts) &
        (df['Name or Item'] == item_name)
    ]
else:
    print("入力に問題がありました。")

df2
```

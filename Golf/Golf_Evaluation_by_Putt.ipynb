```python
import pandas as pd
import numpy as np

path = r"C:\Users\spax2\OneDrive\ドキュメント\Power BI Desktop\Golf\Final\Golf Performance Report Data2 2025-2.xlsx"

df = pd.read_excel(path, sheet_name="Sheet1")

# 型整理
df["Date"] = pd.to_datetime(df["Date"]).dt.date
df["Score of the Hole"] = pd.to_numeric(df["Score of the Hole"], errors="coerce")
df["Putts of the Hole"] = pd.to_numeric(df["Putts of the Hole"], errors="coerce")

# ラウンド単位（1日1ラウンド前提）
rounds = (
    df.groupby(["Date", "Course Name"], as_index=False)
      .agg(
          Score=("Score of the Hole", "sum"),
          Putts=("Putts of the Hole", "sum")
      )
)

rounds = rounds.sort_values("Date").reset_index(drop=True)
df
```

```python
r = rounds["Score"].corr(rounds["Putts"])
r2 = r ** 2

print("相関係数 r =", round(r, 3))
print("パット寄与率 R² =", round(r2, 3))
```

```python
course_r2 = (
    rounds.groupby("Course Name")
          .apply(lambda g: (g["Score"].corr(g["Putts"])**2) if len(g) >= 4 else np.nan)
          .sort_values(ascending=False)
)

print(course_r2)
```

```python
rounds["NonPutts"] = rounds["Score"] - rounds["Putts"]

r_np = rounds["Score"].corr(rounds["NonPutts"])
r2_np = r_np ** 2

print("Non-Putts r =", round(r_np, 3))
print("Non-Putts R² =", round(r2_np, 3))
```

```python
# csvファイルへの書き込み、20250801上手く動いた　但し、mode='w'をmode='a'すると最初の問題が発生

import csv

def add_photo_data(csv_path, photo_id, date, vege_name, tags):

    with open(csv_path, mode='a', newline='', encoding='utf-8-sig') as csvfile:
        # `tags`リストを展開して、各要素を別々のセルに記録
        writer = csv.writer(csvfile)
        row = [photo_id, date, vege_name, *tags] # `*tags`でリスト内の要素を個々のセルに
        writer.writerow(row)
    
    print(f"{photo_id} のデータを追加しました。")
    return True

```

```python
# 写真の日付け入力

from datetime import datetime, timedelta  # timedeltaもインポートする

def input_date():#日付入力の関数
    while True:  # ユーザーが正しい入力をするまでループ
        try:
            year = int(input("年を入力してください（例: 2025）："))
            month = int(input("月を入力してください（例: 5）："))
            day = int(input("日を入力してください（例: 31）："))
            dt = datetime(year, month, day)
            return dt.date()
        except ValueError as e:
            print("無効な日付です。もう一度入力してください：", e)

def main():
    dt_date = input_date()
    if dt_date and dt_date < datetime(2020, 1, 1).date():
        print("2020年1月1日より前の日付は無効です。")
        return None
    return dt_date

if __name__ == "__main__": # main()を実行する
    dt_date = main()  # main() の戻り値を dt_date に代入, dt_dateは外で使える

    if dt_date:
        print("mainの外で表示：", dt_date.strftime("%Y年%m月%d日"))
    else:
        print("無効な日付が入力されたため、処理できません。")

import datetime
dt_input= dt_date

print("撮影日時", dt_input)
```

```python
import csv

# 読み込み元と書き込み先のパス
input_file = r'H:\\農作業\vegetable_garden_photo.csv'
output_file = r'H:\\農作業\vegetable_garden_photo_2.csv'

# Shift_JISで読み込み、utf-8-sigで保存
with open(input_file, mode='r', encoding='shift_jis', newline='') as infile:
    reader = csv.reader(infile)
    rows = list(reader)  # 内容を一旦すべて読み込む

with open(output_file, mode='w', encoding='utf-8-sig', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(rows)

print("変換完了：UTF-8形式で保存されました")
```

```python
import pandas as pd

# CSVファイルを読み込み、タグをリスト形式に変換
file_path = r'H:\農作業\vegetable_garden_photo_2.csv'

def parse_row(row):
    parts = row.split(',')
    return parts[:3], parts[3:]  # 最初の3列とタグを分割

data = []
with open(file_path, mode='r', encoding='utf-8-sig') as file:
    for line in file:
        base, tags = parse_row(line.strip())
        data.append(base + [tags])  # タグをリストとして追加

# DataFrameに変換
columns = ['photo_id', 'date', 'vege_name', 'tags']
df = pd.DataFrame(data, columns=columns)

print(df)

```

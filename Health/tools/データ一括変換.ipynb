```python
import pathlib

# 対象フォルダ
folder = pathlib.Path(r"C:\Users\spax2\Documents\PythonWork")

# 変換ルール
replacements = {
    "血圧（高）": "血圧収縮期",
    "血圧（低）": "血圧拡張期",
}

# 対象ファイル拡張子
target_ext = [".py", ".ipynb"]

for path in folder.glob("**/*"):
    if path.suffix.lower() in target_ext:
        try:
            text = path.read_text(encoding="utf-8")
        except UnicodeDecodeError:
            # 一部のipynbがUTF-8以外の場合はスキップ
            print(f"⚠ 文字コードエラーのためスキップ: {path}")
            continue

        original_text = text
        for old, new in replacements.items():
            text = text.replace(old, new)

        if text != original_text:
            path.write_text(text, encoding="utf-8")
            print(f"✅ 変換完了: {path}")
        else:
            print(f"（変更なし）: {path}")
```

```python
import pathlib
import openpyxl

# 対象フォルダ
folder = pathlib.Path(r"C:\Users\spax2\Documents\PythonWork")

# 変換ルール
replacements = {
    "血圧（高）": "血圧収縮期",
    "血圧（低）": "血圧拡張期",
}

# 対象拡張子
target_ext = [".xlsx"]

for path in folder.glob("**/*"):
    if path.suffix.lower() in target_ext:
        print(f"処理中: {path}")
        try:
            wb = openpyxl.load_workbook(path)
        except Exception as e:
            print(f"⚠ 開けませんでした: {path} ({e})")
            continue

        changed = False

        # 各シート・各セルをチェック
        for ws in wb.worksheets:
            for row in ws.iter_rows():
                for cell in row:
                    if isinstance(cell.value, str):
                        original = cell.value
                        new_value = original
                        for old, new in replacements.items():
                            new_value = new_value.replace(old, new)
                        if new_value != original:
                            cell.value = new_value
                            changed = True

        if changed:
            wb.save(path)
            print(f"✅ 変換＆保存しました: {path}")
        else:
            print(f"（変更なし）: {path}")

        wb.close()

```

```python
import pathlib
import time
import openpyxl

folder = pathlib.Path(r"C:\Users\spax2\Documents\PythonWork")

replacements = {
    "血圧（高）": "血圧収縮期",
    "血圧（低）": "血圧拡張期",
}

target_ext = [".xlsx", ".xlsm"]
MAX_MB = 5  # 大きいファイルは飛ばす（必要なら増やす）

for path in folder.glob("**/*"):
    if path.suffix.lower() in target_ext:
        size_mb = path.stat().st_size / (1024 * 1024)
        if size_mb > MAX_MB:
            print(f"⚠ {size_mb:.1f}MB なのでスキップ: {path}")
            continue

        print(f"\n=== 処理中: {path} ({size_mb:.1f}MB) ===")
        start = time.time()
        try:
            wb = openpyxl.load_workbook(path)
        except Exception as e:
            print(f"⚠ 開けませんでした: {e}")
            continue
        load_sec = time.time() - start
        print(f"  workbook 読み込み時間: {load_sec:.2f} 秒")

        changed = False
        for ws in wb.worksheets:
            for row in ws.iter_rows():
                for cell in row:
                    if isinstance(cell.value, str):
                        original = cell.value
                        new_value = original
                        for old, new in replacements.items():
                            new_value = new_value.replace(old, new)
                        if new_value != original:
                            cell.value = new_value
                            changed = True

        if changed:
            wb.save(path)
            print("  ✅ 変換して保存しました")
        else:
            print("  （変更なし）")

        wb.close()
```
